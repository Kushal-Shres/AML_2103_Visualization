```python
# load data from 4 csv files
df_17_18 = pd.read_csv("1718Pcard.csv")
df_16_17 = pd.read_csv("PCard 1617.csv")
df_15_16 = pd.read_csv("PCard Transactions 15-16.txt", sep=',')
df_14 = pd.read_csv("Purchasing Card Data 2014 v1.csv")
```

```python
# dropping JV date column
df_14.drop(columns=["JV Date"],inplace=True)

# rename columns of first df to be similar to second and third df
df_14.columns = ["Service Area","Account Description","Creditor","Journal Date","Journal Reference","Total"]

# rename columns of fourtht df to be similar to second and third df
df_17_18.columns = ["Transaction Date","Journal Date","Total","Creditor","City","Amount","Country_Code","C_ratio"]
```

```python
# check dimensions of each dataframe
print(df_14.shape,df_15_16.shape,df_16_17.shape,df_17_18.shape)
# print(4142 + 3865 + 4582)
```

```python
# combine first, second and third dataframe as they have same columns
combined_df = pd.concat([df_14, df_15_16,df_16_17], ignore_index=True)
combined_df.dropna(subset=['Journal Date'], inplace=True)

# convert journal date to datetime format
combined_df["JDate"] = combined_df["Journal Date"].apply(lambda x: datetime.strptime(x, "%d/%m/%Y"))
```

```python
# display first 5 data of the combined df
combined_df.head()
```

```python
# replacing values in missing columns Service area and Account Description by comparing to the combined dataframe
for index,item in df_17_18.iterrows():
    try: 
        # only choosing one unique service area
        # same creditor might have multiple service area 
        df_17_18.at[index, "Service Area"] = combined_df[combined_df["Creditor"] == item[3]]["Service Area"].unique()[0]
    except:
        None
    try:
         # only choosing one unique account
        # same creditor might have multiple account 
        df_17_18.at[index, "Account Description"] = combined_df[combined_df["Creditor"] == item[3]]["Account Description"].unique()[0]
    except:
        None
```

```python
# converting date to datetime format in fourth dataframe
df_17_18["JDate"] = df_17_18["Journal Date"].apply(lambda x: datetime.strptime(x, "%d/%m/%y"))
df_17_18.head(3)
```

```python
# joining fourth dataframe to the combined dataframe
df = pd.concat([combined_df, df_17_18],ignore_index=True)
df.shape
```

```python
# Addind a new coulumn quarter according to Jdate which shows which quarter of which year the data falls under
df['quarter'] = df['JDate'].dt.to_period('Q')
```

```python
# replacing comma with empty space and changing the datatype of Total to float
df['Total'] = df['Total'].apply(lambda x: x.replace(",",""))
df['Total'] = df['Total'].astype(float)
```

```python
# grouping the dataset according to the quarter and extracting count and mean of each quarter
quarterly_stats = df.groupby('quarter')['Total'].agg(['count', 'mean']).reset_index()
plt.figure(figsize=(20, 5))

# plotting line graph for means of each quarter
plt.plot(quarterly_stats['quarter'].astype(str), quarterly_stats['mean'], marker='o', color='orange', label='Mean')

# plotting bargraph for count of each quarter
plt.bar(quarterly_stats['quarter'].astype(str), quarterly_stats['count'], color='g', label='Count')

# displaying value for each bar graph
for i, count in enumerate(quarterly_stats['count']):
    plt.text(i, count + 2, str(count), ha='center', va='bottom', fontsize=10)
    
# displayign value for each point in line graph
for i, mean in enumerate(quarterly_stats['mean']):
    plt.text(i, mean + 2, '{:.2f}'.format(mean), ha='center', va='bottom', fontsize=10)

# set title, x label and y label
plt.title('Count and Mean by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Value')
plt.legend()
# plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# grouping data according to service area and quarter and extracting count and mean
grouped_df = df.groupby(['Service Area', 'quarter'])['Total'].agg(['count', 'mean']).reset_index()

# pivot table according to the service areas as columns
pivoted_table = grouped_df.pivot_table(values='count', index='quarter', columns='Service Area', fill_value=0)

# set index and columns for new pivoted dataframe
df_pivot = df.pivot_table(index="quarter", columns='Service Area', values='Total')
df_pivot.head()
```

```python
# plot means for each service area in each quarter
df_pivot.plot(kind='bar', stacked=True, figsize=(15,15))

plt.title('Average spending for each Service Area each Quarter')
plt.xlabel('Quater')
plt.ylabel('Total')
plt.xticks(rotation=90)
plt.legend(title='Service Area')
plt.show()
```

```python
# group data according to service area
service_stats = df.groupby('Service Area')['Total'].agg(['count', 'mean']).reset_index()
```

```python
# plot service area and mean expenditure
plt.figure(figsize=(20, 5))
plt.bar(service_stats['Service Area'].astype(str), service_stats['mean'], color='g', label='Count')
for i, mean in enumerate(service_stats['mean']):
    plt.text(i, mean + 2, '{:.2f}'.format(mean), ha='center', va='bottom', fontsize=10)
plt.xticks(rotation=90)
plt.show()
```

```python
#plot service area and average spending to see the spike in expenditure
plt.figure(figsize=(20, 5))
plt.plot(service_stats['Service Area'].astype(str), service_stats['mean'], marker='o')
plt.xticks(rotation=90)
plt.show()
```

Here we can see that the service areas  CSG Managed Budget,Customer Support Group and Governance have significantly higher spending than others.


```python
# PLotting graphs to see which quarter had significantly more spending service area
pivot_df = df.pivot_table(index="quarter", columns='Service Area', values='Total')

# print(pivot_df.iloc[0:,0:5].head())
# only plotting 10 service areas fro ease to see
pivot_df.iloc[0:,0:10].sort_values(by = "quarter",ascending=False).interpolate(method='linear').plot(kind='line', figsize=(10,5))
plt.title('Average spending for each Service Area each Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

Among the 10 plotted service areas the CSG Managed Budget had a sharp spike in 2014 Q4

```python
# calculating quartiles
Q1 = service_stats['mean'].quantile(0.25)
Q3 = service_stats['mean'].quantile(0.75)
IQR = Q3 - Q1

threshold = Q3 + 1.5 * IQR
spikes = service_stats[service_stats['mean'] > threshold]
print(spikes)
```

Assuming a the cap of the quartiles as threshold for automatic spike detection
Assuming that every Service area with more mean expenditure than the threshold as a spike
we can see that 3 service areas were identified as spike

```python
# using KMeans to try and group the price service area and comments
```

```python
# importing libraries for clustering
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler, LabelEncoder

# grouping the dataframe according to the service area with count and mean
stats = df.groupby('Service Area')['Total'].agg(['count', 'mean']).reset_index()
# creating a new stats df
stats_array = pd.DataFrame(stats.values,columns=["Service Area","count","mean"])

# initializing label encoder
label_encoder = LabelEncoder()

# Adding new column to store encoded Service Area
stats_array["Service_Area_code"] = label_encoder.fit_transform(stats_array["Service Area"])
# dropping service area column
encoded_data = stats_array.drop(columns=["Service Area"])

# initializing k means 
kmeans = KMeans(n_clusters=3, random_state=42)

# adding new cluster column obtained from k means algorithm
stats_array['Cluster'] = kmeans.fit_predict(encoded_data)

# Assigning cluster colors
cluster_colors = sns.color_palette('Paired', len(stats_array['Cluster'].unique()))

# plotting a scatters plot 
sns.scatterplot(data=stats_array, x='count', y='mean', hue='Cluster', palette=cluster_colors)
plt.xlabel('Number of Transactions')
plt.ylabel('Total Amount')
plt.title('Clustering of Service Areas based on Spending Behavior')
plt.show()
```

Here we can see that the data points are divided into 3 clusters

```python
# using Isolation forest to try and find anamolies
```

```python
# Importing library for isolation Forest
from sklearn.ensemble import IsolationForest

# encoding the service area into new column
label_encoder = LabelEncoder()
df['service_area_encoded'] = label_encoder.fit_transform(df['Service Area'])

# Selecting features for anomaly detection
features = ['service_area_encoded', 'Total']  

# Initialize Isolation Forest
model = IsolationForest(contamination=0.05)

# Fit the model on the data
model.fit(df[features])

# Predict outliers/anomalies
df['anomaly_score'] = model.decision_function(df[features])
df['anomaly'] = model.predict(df[features])

plt.figure(figsize=(10, 6))
sns.histplot(df['anomaly_score'], kde=True, bins=20, color='blue')
plt.title('Distribution of Anomaly Scores in Sample Transactions')
plt.xlabel('Isolation Score')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()

# Display top Anamolies
count = 0
grouped_anomalies = df[df['anomaly'] == -1].groupby('Service Area')

for service_area, group in grouped_anomalies:
    if(count<5):

        print(f"Anomalies for Service Area: {service_area}")
        top_anomalies = group.head(5)
        print(top_anomalies)
        print()
        count+=1
```
