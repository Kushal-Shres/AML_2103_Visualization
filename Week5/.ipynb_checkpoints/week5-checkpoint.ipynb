```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv("CustomersTransactions2021.csv")
```

```python
df.head()
```

```python
#checking if id is sorted or not
df[["id"]]
```

```python
#sorting the column
df.sort_values(by = 'id', inplace = True) #inplace = true make sure it reflects on the database
```

### Line Graph

```python
plt.plot(df['id'], df['txn_total'])
```

```python
#Naming the axis
plt.title("Line Graph of Transaction Total")
plt.xlabel("IDS")
plt.ylabel("Txn Amount on $", rotation = 0, labelpad=40)#to rotate the labels

plt.plot(df['id'], df['txn_total'], 'go--', label= 'Transaction')
plt.legend()
```

```python
# Making the line Magenta Colour and use asterisk as markers
plt.title("Line Graph of Transaction Total")
plt.xlabel("IDS")
plt.ylabel("Txn Amount on $", rotation = 0, labelpad=40)#to rotate the labels
#
plt.plot(df['id'], df['txn_total'], 'm--*', label= 'Transaction')
plt.legend()
```

```python
# SHowing that (x) is optional in plt.plot
plt.plot([10, 20, 30, 40], 'm*:')
```

```python
# Annotating every marker: 
plt.figure(figsize=(15,5))
plt.title("Line Graph of Transaction Total")
plt.xlabel("IDS")
plt.ylabel("Txn Amount on $", rotation = 0, labelpad=40)#to rotate the labels
#
plt.plot(df['id'], df['txn_total'], 'm--*', label= 'Transaction')
plt.legend()

for x, y in zip(df['id'], df['txn_total']):
    label= "{:.2f}".format(y)
    plt.annotate(label, (x, y))
```

### Histogram

```python
# Drawing a histogram

plt.hist(df['txn_total'], bins=5, alpha=0.45, color='red', edgecolor= 'black') 

plt.show() 
```

```python
plt.figure(figsize=(15,6))
plt.hist(df['txn_total'],bins=5, color='orange', edgecolor='white', label='values')

plt.xlabel('Value', labelpad=40, rotation=30)
plt.ylabel('frequency',labelpad=40, rotation=30)
plt.title('Histogram for Total transactions in $')
plt.legend()
```

```python
observation, bins, num_bins, = plt.hist(df['txn_total'], label= 'TXN_TOTAL', bins= 5)
plt.xlabel("Txn amount in $")
plt.ylabel('Num okf observations')
plt.title('Histogram of  Txn_total')
plt.legend()
```
