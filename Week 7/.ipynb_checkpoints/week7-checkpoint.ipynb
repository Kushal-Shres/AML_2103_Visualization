```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv("CustomersTransactions2021.csv")
```

```python
#checking if id is sorted or not
df[["id"]]
```

```python
#sorting the column
df.sort_values(by = 'id', inplace = True) #inplace = true make sure it reflects on the database
```

### Line Graph

```python
plt.plot(df['id'], df['txn_total'])
```

```python
#Naming the axis
plt.title("Line Graph of Transaction Total")
plt.xlabel("IDS")
plt.ylabel("Txn Amount on $", rotation = 0, labelpad=40)#to rotate the labels

plt.plot(df['id'], df['txn_total'], 'go--', label= 'Transaction')
plt.legend()
```

```python
# Making the line Magenta Colour and use asterisk as markers
plt.title("Line Graph of Transaction Total")
plt.xlabel("IDS")
plt.ylabel("Txn Amount on $", rotation = 0, labelpad=40)#to rotate the labels
#
plt.plot(df['id'], df['txn_total'], 'm--*', label= 'Transaction')
plt.legend()
```

```python
# SHowing that (x) is optional in plt.plot
plt.plot([10, 20, 30, 40], 'm*:')
```

```python
# Annotating every marker: 
plt.figure(figsize=(15,5))
plt.title("Line Graph of Transaction Total")
plt.xlabel("IDS")
plt.ylabel("Txn Amount on $", rotation = 0, labelpad=40)#to rotate the labels
#
plt.plot(df['id'], df['txn_total'], 'm--*', label= 'Transaction')
plt.legend()

for x, y in zip(df['id'], df['txn_total']):
    label= "{:.2f}".format(y)
    plt.annotate(label, (x, y))
```

# Bar Graphs

```python
plt.bar(df['Category'].unique(), df['Category'].value_counts(), color=['blue', 'orange', 'green'])

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Counts of Categories')

# Show the plot
plt.show()
```

```python
# Aggregate the data by category and calculate total amount spent per category
total_amount_per_category = df.groupby('Category')['txn_total'].sum()

# Create bar graph
plt.bar(total_amount_per_category.index, total_amount_per_category.values, color=['blue', 'orange', 'green'])

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Total Amount Spent')
plt.title('Total Amount Spent per Category')

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Group by category and payment method, then sum the amounts
grouped = df.groupby(['Category', 'txn_type']).aggregate({'txn_total':'sum'}).unstack(fill_value=0)

# Plot stacked bar chart
grouped.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Total Amount Spent')
plt.title('Total Amount Spent per Category and Payment Method')

# Show the plot
plt.legend(title='Payment Method')
plt.xticks(rotation=45)
plt.show()
```

# Bar graphs

 Simple Bar Graph:
1. vertical Bar graph are vertically orinted 
    *x=**list of** centres of bars
    *heights= list of heights of bars
    *width= single_value Default=0.8
2. Horizontal bar graph 
    plt.barh(....)

```python
cat = ['A', 'B', 'C', 'D']
heights = [10, 20, 50, 15]

# Create the bar chart
plt.bar(cat, heights, color='maroon', width= 0.4)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Height')
plt.title('Bar Chart of Categories')
```

```python
categories = ['A', 'B', 'C', 'D']
heights = [10, 20, 50, 15]

# Create the bar chart
plt.barh(categories, heights, color='blue')

# Add labels and title
plt.xlabel('Height')
plt.ylabel('Categories')
plt.title('Bar Chart of Categories')
```
